```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
hws_all, hwos_all = [], []
with open("all.txt") as inp:
    inp.readline()
    inp.readline()
    
    for line in inp:
        hws_all.append(int(line[:-1].split(' ')[1]))
        hwos_all.append(int(line[:-1].split(' ')[2]))
        
hws_1, hwos_1 = [], []
with open("1.txt") as inp:
    inp.readline()
    inp.readline()
    
    for line in inp:
        hws_1.append(int(line[:-1].split(' ')[1]))
        hwos_1.append(int(line[:-1].split(' ')[2]))
        
hws_2, hwos_2 = [], []
with open("2.txt") as inp:
    inp.readline()
    inp.readline()
    
    for line in inp:
        hws_2.append(int(line[:-1].split(' ')[1]))
        hwos_2.append(int(line[:-1].split(' ')[2]))
        
hws_3, hwos_3 = [], []
with open("3.txt") as inp:
    inp.readline()
    inp.readline()
    
    for line in inp:
        hws_3.append(int(line[:-1].split(' ')[1]))
        hwos_3.append(int(line[:-1].split(' ')[2]))
        
hws_4, hwos_4 = [], []
with open("4.txt") as inp:
    inp.readline()
    inp.readline()
    
    for line in inp:
        hws_4.append(int(line[:-1].split(' ')[1]))
        hwos_4.append(int(line[:-1].split(' ')[2]))
        
hws_5, hwos_5 = [], []
with open("5.txt") as inp:
    inp.readline()
    inp.readline()
    
    for line in inp:
        hws_5.append(int(line[:-1].split(' ')[1]))
        hwos_5.append(int(line[:-1].split(' ')[2]))
```

```python
plt.figure(figsize=(20, 5))

plt.title("h-index -- with vs without self citations")

ax1 = plt.subplot(1, 2, 1)
ax1.set_xlabel("h-index with self citations")
ax1.set_ylabel("h-index without self citations")

ax1.scatter(hws_all, hwos_all, alpha=0.02)

ax1.set_xlim(10, 60)
ax1.set_ylim(10, 60)

ax2 = plt.subplot(1, 2, 2)
ax2.set_xlabel("h-index with self citations")
ax2.set_ylabel("h-index without self citations")

ax2.scatter(hws_1, hwos_1, c="k", alpha=0.005)
ax2.scatter(hws_2, hwos_2, c="b", alpha=0.005)
ax2.scatter(hws_3, hwos_3, c="g", alpha=0.005)
ax2.scatter(hws_4, hwos_4, c="c", alpha=0.005)
ax2.scatter(hws_5, hwos_5, c="m", alpha=0.005)

ax2.set_xlim(10, 60)
ax2.set_ylim(10, 60)

plt.show()
```

```python
hws_all_neq, hwos_all_neq = [hws_all[i] for i in range(len(hws_all)) if hws_all[i] != hwos_all[i]], [hwos_all[i] for i in range(len(hws_all)) if hws_all[i] != hwos_all[i]]

hws_1_neq, hwos_1_neq = [hws_1[i] for i in range(len(hws_1)) if hws_1[i] != hwos_1[i]], [hwos_1[i] for i in range(len(hws_1)) if hws_1[i] != hwos_1[i]]
hws_2_neq, hwos_2_neq = [hws_2[i] for i in range(len(hws_2)) if hws_2[i] != hwos_2[i]], [hwos_2[i] for i in range(len(hws_2)) if hws_2[i] != hwos_2[i]]
hws_3_neq, hwos_3_neq = [hws_3[i] for i in range(len(hws_3)) if hws_3[i] != hwos_3[i]], [hwos_3[i] for i in range(len(hws_3)) if hws_3[i] != hwos_3[i]]
hws_4_neq, hwos_4_neq = [hws_4[i] for i in range(len(hws_4)) if hws_4[i] != hwos_4[i]], [hwos_4[i] for i in range(len(hws_4)) if hws_4[i] != hwos_4[i]]
hws_5_neq, hwos_5_neq = [hws_5[i] for i in range(len(hws_5)) if hws_5[i] != hwos_5[i]], [hwos_5[i] for i in range(len(hws_5)) if hws_5[i] != hwos_5[i]]
```

```python
plt.figure(figsize=(20, 5))

plt.title("h-index -- with vs without self citations")

ax1 = plt.subplot(1, 2, 1)
ax1.set_xlabel("h-index with self citations")
ax1.set_ylabel("h-index without self citations")

ax1.scatter(hws_all_neq, hwos_all_neq, alpha=0.05)

ax1.set_xlim(10, 60)
ax1.set_ylim(10, 60)

ax2 = plt.subplot(1, 2, 2)
ax2.set_xlabel("h-index with self citations")
ax2.set_ylabel("h-index without self citations")

ax2.scatter(hws_1_neq, hwos_1_neq, c="k", alpha=0.05)
ax2.scatter(hws_2_neq, hwos_2_neq, c="b", alpha=0.05)
ax2.scatter(hws_3_neq, hwos_3_neq, c="g", alpha=0.05)
ax2.scatter(hws_4_neq, hwos_4_neq, c="c", alpha=0.05)
ax2.scatter(hws_5_neq, hwos_5_neq, c="m", alpha=0.05)

ax2.set_xlim(10, 60)
ax2.set_ylim(10, 60)

plt.show()
```

```python
hdiff_all = [(hws_all[i] - hwos_all[i]) / hws_all[i] for i in range(len(hws_all))]

hdiff_1 = [(hws_1[i] - hwos_1[i]) / hws_1[i] for i in range(len(hws_1))]
hdiff_2 = [(hws_2[i] - hwos_2[i]) / hws_2[i] for i in range(len(hws_2))]
hdiff_3 = [(hws_3[i] - hwos_3[i]) / hws_3[i] for i in range(len(hws_3))]
hdiff_4 = [(hws_4[i] - hwos_4[i]) / hws_4[i] for i in range(len(hws_4))]
hdiff_5 = [(hws_5[i] - hwos_5[i]) / hws_5[i] for i in range(len(hws_5))]

print("All:\t\t\t\t\t\tMean: {:.6f}\t\tStDev: {:.6f}\t\tNonZeroPc: {:8.6f}".format(np.mean(hdiff_all), np.std(hdiff_all), (len(hdiff_all) - hdiff_all.count(0)) / len(hdiff_all)))

print("# 1:\t\t\t\t\t\tMean: {:.6f}\t\tStDev: {:.6f}\t\tNonZeroPc: {:8.6f}".format(np.mean(hdiff_1), np.std(hdiff_1), (len(hdiff_1) - hdiff_1.count(0)) / len(hdiff_1)))
print("# 2:\t\t\t\t\t\tMean: {:.6f}\t\tStDev: {:.6f}\t\tNonZeroPc: {:8.6f}".format(np.mean(hdiff_2), np.std(hdiff_2), (len(hdiff_2) - hdiff_2.count(0)) / len(hdiff_2)))
print("# 3:\t\t\t\t\t\tMean: {:.6f}\t\tStDev: {:.6f}\t\tNonZeroPc: {:8.6f}".format(np.mean(hdiff_3), np.std(hdiff_3), (len(hdiff_3) - hdiff_3.count(0)) / len(hdiff_3)))
print("# 4:\t\t\t\t\t\tMean: {:.6f}\t\tStDev: {:.6f}\t\tNonZeroPc: {:8.6f}".format(np.mean(hdiff_4), np.std(hdiff_4), (len(hdiff_4) - hdiff_4.count(0)) / len(hdiff_4)))
print("# 5:\t\t\t\t\t\tMean: {:.6f}\t\tStDev: {:.6f}\t\tNonZeroPc: {:8.6f}".format(np.mean(hdiff_5), np.std(hdiff_5), (len(hdiff_5) - hdiff_5.count(0)) / len(hdiff_5)))
```

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.title("Probability Density Function of h-index with Self Citations")

kde = scipy.stats.gaussian_kde(hws_1)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 1')
kde = scipy.stats.gaussian_kde(hws_2)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 2')
kde = scipy.stats.gaussian_kde(hws_3)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 3')
kde = scipy.stats.gaussian_kde(hws_4)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 4')
kde = scipy.stats.gaussian_kde(hws_5)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 5')

plt.legend(loc='center')

plt.subplot(1, 2, 2)
plt.title("Probability Density Function of h-index without Self Citations")

kde = scipy.stats.gaussian_kde(hwos_1)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 1')
kde = scipy.stats.gaussian_kde(hwos_2)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 2')
kde = scipy.stats.gaussian_kde(hwos_3)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 3')
kde = scipy.stats.gaussian_kde(hwos_4)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 4')
kde = scipy.stats.gaussian_kde(hwos_5)
plt.plot(np.linspace(10, 30), kde(np.linspace(10, 30)), label='Community # 5')

plt.legend(loc='center')

plt.show()
```
